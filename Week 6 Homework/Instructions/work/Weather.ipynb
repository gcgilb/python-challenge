```python
# Dependencies
import json
import requests
import pandas as pd
from citipy import citipy
from config import api_key
import matplotlib.pyplot as plt
import random
```

```python
cities = []
city1 = []

for x in range(1500):
    
    lat =(random.randint(-90.00,90.00))
    lon =(random.randint(-180,180))
    
    cities = citipy.nearest_city(lat, lon)
    city1.append(cities.city_name)
                
```

```python
# Save config information.
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
no_dupes = [x for n, x in enumerate(city1) if x not in city1[:n]]
#print(no_dupes)
```

```python
# Loop through the list of cities and perform a request for data on each
for city in no_dupes:
   
    try:    
        response = requests.get(query_url + city).json()
        lat.append(response['coord']['lat'])
        temp.append(response['main']['temp_max'])
        hum.append(response['main']['humidity'])
        cloud.append(response['clouds']['all'])
        wind.append(response['wind']['speed'])
        date.append(response['dt'])
        name.append(response['name'])
        lon.append(response['coord']['lon'])
        cty.append(response['sys']['country'])
        number=(response['id'])
        
        print(city, number)
       
        
    except (KeyError):
        
        count + 1      
```

```python
# create a data frame from cities, lat, and temp
weather = {
    "City": name,
       "Lat": lat,
    "Max Temp": temp,
    "Cloudiness": cloud,
    "Wind Speed": wind,
    "Humidity": hum,
    "Country": cty,
    "Date": date,
    "Lng": lon
    
    }
weather_data = pd.DataFrame(weather)
weather_data.head()
```
